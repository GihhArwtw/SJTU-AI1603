综合练习1：从键盘读入一组字符串（字符串只包含大小写字母），字符串用空格分隔。由于这组字符串是人名，请把每个单词的首字母大写，其他字母小写。然后按照首字母的ASCII码的顺序降序输出。（3分）

```python
# 综合练习1
strings = input()

# TODO
def convert_strings(strings):
    # TODO
    
    name = list(map(str,strings.split()))
    name.sort(reverse = True)
    
    for i in range(len(name)):
        name[i] = chr(ord(name[i][0])-32) + name[i][1:]
    
    result = name[0]
    for i in range(1,len(name)):
        result = result + ", " + name[i]
    
    return result
# end convert_strings


'''
Input: strings = 'zhao qian sun li'
Expected Output: Zhao, Sun, Qian, Li

'''

#Test
print(convert_strings(strings))
```

综合练习2:设计一个函数，可用于自动判断二进制、八进制、十六进制输入，并转换成十进制。可以使用内置函数实现（3分）。自定义的进制转换实现（1分）。

```python
# 综合练习2

# TODO
# Hint: Checkout Built-in Functions 'bin', 'hex', 'oct', 'int' in Python Offical Docs
def convert_binary2_bulitin(num):
    result = int(num,0)
    return result
# end convert_binary2_builtin


def convert_binary2(num):
    def value(bit):
        if bit>='0' and bit<='9' : return eval(bit)
        else:                      return ord(bit)-87
    # end value
        
    def convert(num, base):
        sum = value(num[0])
        for i in range(1,len(num)):
            sum = sum*base + value(num[i])
        return sum
    # end convert
        
    if (num[1] == 'b'):    result = convert(num[2:],2)         # binary
    elif (num[1] == 'x'):  result = convert(num[2:],16)        # hexadecimal
    elif (num[1] == 'o'):  result = convert(num[2:],8)         # octal
    else:                  result = num                        # decimal
        
    return result
# end convert_binary2

'''
Inputs: '0b111', '0xaaa',  '0o1237'
Expected Outputs: 7, 2730, 671
'''

# Test

n = input()
print(convert_binary2_bulitin(n))
print(convert_binary2(n))
```

综合练习3: 设计一个装饰器。要求作用在任何函数上时都可以得到这个函数的执行时间。装饰器的输出格式如下：“A executed in B ms”。其中A为被装饰函数的名字，B为时间。(3分)

```python
# Hint, import time, functools before the implementation 

def metric(fn):
    import functools as ft
    @ft.wraps(fn)
    def timer(*val,**par):
        import time
        start = time.time()                  # the unit of time.time() is "s"
        result = fn(*val,**par)
        end = time.time()
        print("test executed in %s ms" % ((end-start)*1000))
        return result
    return timer
#end metric

@metric
def test(x):
    for i in range(x):
        continue

test(100000)
test(19000)

'''
Expected Outputs:
test executed in 0.0071620941162109375 ms
test executed in 0.0013530254364013672 ms
'''
```
